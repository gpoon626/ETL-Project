```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import pyspark
import findspark
findspark.init()
from pyspark.sql import SparkSession

%matplotlib inline
```

```python
df1 = pd.read_csv('Genre_Table_No_Index.csv')
df2 = pd.read_csv('Artist_Table_No_Commas.csv')
df3 = pd.read_csv('Songtitle_No_Commas.csv')
master_df = pd.read_csv('Spotify-2000.csv')
```

```python
df3.plot.scatter(x='Year', y='Popularity', c='red', s=50, figsize=(12,6))
```

- Spotify was founded in 2006
- From the scatterplot, we can see that our data sample is greater from 1990s onwards
- There should be unpopular songs (<20) across the years however none appear before the mid 90s
- We can infer that the older less popular songs never made it on the music platform as they did not survive the test of time

```python
fig, ax = plt.subplots(figsize=(10,10))
pop_corr = df3.pivot_table(values='Popularity', index='Danceability', columns='Year')
ax = sns.heatmap(pop_corr,cmap='coolwarm')
ax.invert_yaxis()
```

- With the heatmap, we can see that the contrast in color was greater as the years go by which could mean that there is a discrepancy between the popularity of songs and the year they were released
- darker colors appear more often after the 90s
- from 1950 -> 1960 there were little to no unpopular songs, which means they likely never made it onto the spotify platform
- We can also see that there is a correlation between a song's danceability and it's popularity (there are more red colors when >70 danceability)
